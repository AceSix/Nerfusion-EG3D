```python
import matplotlib.pyplot as plt
```

```python
loss = []
step = []
with open('script_train_ae.o2532296', 'r') as f:
    lines = f.readlines()
    for line in lines:
        seps = line.strip().replace('[', '').replace(']', '').split('-')
        print(seps)
#         print(int(seps[1]), float(seps[5]).split)
        loss.append(float(seps[6]))
        step.append(int(seps[1]))
```

```python
plt.plot(step, loss)
plt.ylim((0, 70))
plt.ylabel('MSE loss')
plt.ylabel('step')
plt.show()
```
